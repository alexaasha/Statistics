```python
import numpy as np
from matplotlib import pyplot as plt
```

Creating dataset

```python
data = np.random.poisson(5, 10000)
# np.random.randn(1000)

# Freedman-Diaconis
q25, q75 = np.percentile(data, [25, 75])
bin_width = 2 * (q75 - q25) * len(data) ** (-1/3)
bins = round((data.max() - data.min()) / bin_width)

hist = plt.hist(data, bins=bins)
plt.show()
```

Mode

```python
plt.hist(data, bins=bins)
plt.axvline(x=hist[1][np.argmax(hist[0])], color='r', linestyle ='--')
plt.show()
```

Median

```python
plt.hist(data, bins=bins)
sorted_data = sorted(data)
center = sorted_data[len(data) // 2] if len(data) % 2 == 0 \
else (sorted_data[len(data) // 2] + sorted_data[len(data) // 2 + 1]) // 2

plt.axvline(x=center, color='r', linestyle ='--')
plt.show()
```

Mean

```python
plt.hist(data, bins=bins)
plt.axvline(x=data.mean(), color='r', linestyle ='--')
plt.show()
```
