```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
size_of_set = 100
x = np.random.randint(low=0, high=100, size=size_of_set)
y = np.random.randint(low=0, high=100, size=size_of_set)

plt.plot(x, y, 'bo')
plt.show()
```

Perform simple random sample

```python
num_of_samples = 15
rnd_ind = np.random.randint(low=0, high=100, size=num_of_samples)

x_sample = [x[i] for i in rnd_ind]
y_sample = [y[i] for i in rnd_ind]

plt.plot(x, y, 'bo')
plt.plot(x_sample, y_sample, 'ro')
plt.show()
```

Perform cluster sample. Let's divide set into 4 groups

```python
delimiter = size_of_set // 2
h_delim = np.full((size_of_set, ), delimiter)

plt.plot(x, y, 'bo')
plt.plot(h_delim, 'r--')
a = plt.axvline(x=delimiter, color='r', linestyle ='--')
plt.show()

offset = 25 # to distinguish clusters

cluster1_ind = np.intersect1d(np.where(x < delimiter), np.where(y < delimiter))
cluster1 = (x[cluster1_ind] - offset, y[cluster1_ind] - offset)

cluster2_ind = np.intersect1d(np.where(x > delimiter), np.where(y < delimiter))
cluster2 = (x[cluster2_ind] + offset, y[cluster2_ind] - offset)

cluster3_ind = np.intersect1d(np.where(x > delimiter), np.where(y > delimiter))
cluster3 = (x[cluster3_ind] + offset, y[cluster3_ind] + offset)

cluster4_ind = np.intersect1d(np.where(x < delimiter), np.where(y > delimiter))
cluster4 = (x[cluster4_ind] - offset, y[cluster4_ind] + offset)

plt.plot(cluster1[0], cluster1[1], 'bo')
plt.plot(cluster2[0], cluster2[1], 'ro')
plt.plot(cluster3[0], cluster3[1], 'go')
plt.plot(cluster4[0], cluster4[1], 'yo')
plt.show()
```
